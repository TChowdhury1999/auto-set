```python
def fix_data(x):
    """ For get_start_end to work, you need the general starting position to have a low value before oscillations begin
        Some exercises this is not the case so we simply have to flip the data
        This function detects if this is needed and then applies the flip
    """
    
    # get average value of first few data points
    starting_avg = np.mean(x[0:10])
    
    # now get the maximum and minimum values
    max_val = max(x)
    min_val = min(x)
    
    # the desired situation is that max - avg > avg - min
    if (max_val - starting_avg) < (starting_avg - min_val):
        return -x
    else:
        return x
    
```

```python
def get_reps(x, height_tolerance=0.5, period_tolerance = 0.5):
    """
    Get the peaks and troughs of oscillations in array x
    Tolerance is how leniant to accept peaks
    """
    
    # make sure the starting position is at the bottom
    x = fix_data(x)
    
    # get the max of x for use later
    max_x = max(x)
    
    # compute autocorrelation of x and restrict to positive lags
    corr = np.correlate(x, x, mode='full')[len(x)-1:]
    
    
    # get the first peak of this which would be the period of oscillations
    peaks, _ = find_peaks(corr)
    period = peaks[0]
    
    # find the peaks in the signal
    rep_peaks, _ = find_peaks(x, height = max_x*height_tolerance, distance = period * period_tolerance)
    
    # now find the troughs on smoothed data
    troughs, _ = find_peaks(-savgol_filter(x, 10, 2))
    
    # the start will be the first trough before the first peak and end first trough after last peak
    min_trough = np.max(troughs[troughs < np.min(rep_peaks)])
    max_trough = np.min(troughs[troughs > np.max(rep_peaks)])

    rep_troughs = troughs[(troughs >= min_trough) & (troughs <= max_trough)]
    
    
    return (rep_peaks, rep_troughs)
```

```python
get_reps(x)
```

```python
def pretty_plot(x):
    """ Takes the pose data and plots a nice graph only of the oscillating portion with 
        eccentric and concentric regions and times labelled
    """
    
    # lets get the reps
    rep_peaks, rep_troughs = get_reps(x)
    
    # lets restrict the data only to the rep portion
    x = x[rep_troughs[0]-1:rep_troughs[-1]+1]
    
    # now lets smooth out the data
    moving_avg = savgol_filter(x, 10, 2)
    
    plt.plot(moving_avg)
```

```python
pretty_plot(x)
```
